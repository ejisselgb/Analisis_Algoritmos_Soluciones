```python
import time
import numpy.random as nprnd


## Declarqacion de la funcion - Ordenamiento burbuja
def burbuja(array, show):
    arrayb = list(array)
    for i in range(1,len(arrayb)):
        for j in range(len(arrayb)-i):
            if arrayb[j] > arrayb[j+1]:
                aux = arrayb[j]
                arrayb[j] = arrayb[j+1]
                arrayb[j+1] = aux
    if show == True: print("Arreglo ordenado - Burbuja: ", arrayb)


## Declaracion de la funcion - Ordenamiento por seleccion
def seleccion(array, show):
    arrays = list(array)
    for i in range(len(arrays)-1):
        minimo = i
        for j in range(i+1,len(arrays)):
            if arrays[j] < arrays[minimo] :
                minimo = j
        aux = arrays[i]
        arrays[i] = arrays[minimo]
        arrays[minimo] = aux
    if show == True: print("Arreglo ordendo - Seleccion: ", arrays)


## Declaracion de la funcion - Ordenamiento por insercion
def insercion(array, show):
    arrayi = list(array)
    for i in range(len(arrayi)):
        for j in range(i,0,-1):
            while (j > 0) & (arrayi[j-1] > arrayi[j]) :
                aux = arrayi[j]
                arrayi[j] = arrayi[j-1]
                arrayi[j-1] = aux
    if show == True: print("Arreglo ordenado - Insercion: ", arrayi)
    


## Declaracion de la funcion - Ordenamiento shell
def shell(array, show):
    arraysh = list(array)
    intervalo = len(arraysh)/2
    while intervalo > 0 :
        for i in range(intervalo, len(arraysh)):
            j = i
            aux = arraysh[i]
            while j >= intervalo and arraysh[j-intervalo] > aux :
                arraysh[j] = arraysh[j-intervalo]
                j = j - intervalo
                arraysh[j] = aux
        if intervalo == 2:
            intervalo = 1
        else:
            intervalo = intervalo/2
    if show == True: print("Arreglo ordenado - Shell: ", arraysh)
    

    
## Declaracion de la funcion - Ordenamiento rapido - Quicksort    
def quicksort(array, izq, der, show):
    i = izq
    j = der
    arrayQuick=array[(izq + der)/2]
    while( i <= j ):
        while array[i]<arrayQuick and j<=der:
            i=i+1
        while arrayQuick<array[j] and j>izq:
            j=j-1
        if i<=j:
            aux = array[i]; array[i] = array[j]; array[j] = aux;
            i=i+1;  j=j-1;
        if izq < j:
            quicksort( array, izq, j, False )
    if i < der:
        quicksort( array, i, der, False )
    if show == True: print("Arreglo ordenado - Quicksort: ", array)
        

## Llamado de las funciones - calcular tiempo de ejecucion
def calculateTime(method, array, show):
    t0 = time.time()
    method(array,show)
    t= time.time()-t0
    return t


## Creacion de arreglo - valores aleatorios entre 0 y 200 - len del arreglo 100
def generateArray():
    array1 = nprnd.randint(200, size=200)
    array2 = nprnd.randint(400, size=600)
    array3 = nprnd.randint(900, size=1000)
    return array1, array2, array3


array1, array2, array3 = generateArray() # Asignacion de valores a los arreglos
arrayq1 = list(array1)
arrayq2 = list(array2)


tb1 = calculateTime(burbuja, array1, False)
ts1 = calculateTime(seleccion, array1, False)
ti1 = calculateTime(insercion, array1, False)
tsh1 = calculateTime(shell, array1, False)


tq0 = time.time()
quicksort(arrayq1, 0, len(array1)-1, False)
tq1= time.time()-tq0

tqs0 = time.time()
quicksort(arrayq2, 0, len(array2)-1, False)
tqs1= time.time()-tqs0

tb2 = calculateTime(burbuja, array2, False)
ts2 = calculateTime(seleccion, array2, False)
ti2 = calculateTime(insercion, array2, False)
tsh2 = calculateTime(shell, array2, False)
```

```python
## Grafica rendimiento algoritmos de ordenamiento

import pylab

x=[len(array1),len(array2)]
ttb1=[tb1, tb2]
tts1=[ts1, ts2]
tti1=[ti1, ti2]
ttsh1=[tsh1, tsh2]
ttq1=[tq1, tqs1]

pylab.plot(ttb1, x, label = "Burbuja")
pylab.plot(tts1, x, label = "Seleccion")
pylab.plot(tti1, x, label = "Insercion")
pylab.plot(ttsh1, x, label = "Shell")
pylab.plot(ttq1, x, label = "Quicksort")
pylab.legend(loc='upper right')

pylab.xlabel("Tiempo")
pylab.ylabel("Datos") 

pylab.show()
```
